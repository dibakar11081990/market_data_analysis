```python
import snowflake.connector
import polars as pl

class ProcessInSnowflake:
    def __init__(self):

        # Snowflake connection details
        self.account = 'tata.us-east-1'
        self.user = 'user'
        self.password = 'pwd'
        self.database = 'DEV'
        self.schema = 'MQA_AMPLIFY_ABM'
        self.warehouse = 'BSM_ENGINEERING_WAREHOUSE_MEDIUM'
        self.role = 'BSM_ENGINEERING'

        # Establish connection
        self.conn = snowflake.connector.connect(
                    account=self.account,
                    user=self.user,
                    password=self.password,
                    warehouse=self.warehouse,
                    DATABASE=self.database,   
                    SCHEMA=self.schema,      
                    role=self.role)
        
        

    def snowflake_execution(self,query:str):
        
        try:
            df = pl.read_database(query, connection=self.conn)
        except Exception as e:
            print(f"General Error occurred: {e}")
            return "Query Error"
        
        if df.is_empty():
            print("Query executed successfully but returned no results.")
            return "Pass"
        else:
            return "Fail"
```
