```python
import os

def rename_files_to_lowercase(folder_path):
    """
    Renames all files in the given folder to lowercase.
    
    Parameters:
        folder_path (str): Path to the folder.
    """
    for filename in os.listdir(folder_path):
        original_path = os.path.join(folder_path, filename)

        # Skip if it's not a file
        if not os.path.isfile(original_path):
            continue

        lowercase_name = filename.lower()
        lowercase_path = os.path.join(folder_path, lowercase_name)

        if original_path != lowercase_path:
            os.rename(original_path, lowercase_path)
            print(f"Renamed: {filename} -> {lowercase_name}")
        else:
            print(f"Skipped (already lowercase): {filename}")

# Example usage
folder = "C:/Users/mandald/OneDrive - tredence_analytics/gitFolder/marketing-data-analytics/dags/dbt/mars_dbt/models/activity_marketo_v2_driver"
rename_files_to_lowercase(folder)
```
